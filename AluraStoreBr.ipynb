### Importação dos dados



```python
import pandas as pd

url = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_1.csv"
url2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_2.csv"
url3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_3.csv"
url4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_4.csv"

loja = pd.read_csv(url)
loja2 = pd.read_csv(url2)
loja3 = pd.read_csv(url3)
loja4 = pd.read_csv(url4)

loja.head()
```

#1. Análise do faturamento


```python
lojas_dfs = {
    'Loja 1': loja,
    'Loja 2': loja2,
    'Loja 3': loja3,
    'Loja 4': loja4
}
```

```python
faturamentos = {}
coluna_faturamento = 'Preço'
```

```python
print("\n--- Calculando Faturamento de Cada Loja (baseado na coluna 'Preço') ---")
for nome_loja, df in lojas_dfs.items():
    if df is not None and not df.empty:
        if coluna_faturamento in df.columns:
            faturamento_loja = df[coluna_faturamento].sum()
            faturamentos[nome_loja] = faturamento_loja
            print(f"Faturamento da {nome_loja}: R$ {faturamento_loja:,.2f}")
        else:
            print(f"AVISO: Coluna '{coluna_faturamento}' não encontrada na {nome_loja}. Faturamento não calculado.")
            faturamentos[nome_loja] = 0 # Ou None, ou pular
    else:
        print(f"AVISO: DataFrame para {nome_loja} está vazio ou não foi carregado.")
        faturamentos[nome_loja] = 0
```

```python
df_faturamento_total = pd.DataFrame(list(faturamentos.items()), columns=['Loja', 'Faturamento (Preço)'])
df_faturamento_total = df_faturamento_total.sort_values(by='Faturamento (Preço)', ascending=False)

print("\n--- Resumo do Faturamento por Loja (Ordenado) ---")
print(df_faturamento_total)

# Calcular o faturamento total de todas as lojas
faturamento_geral = df_faturamento_total['Faturamento (Preço)'].sum()
print(f"\nFaturamento Geral (todas as lojas): R$ {faturamento_geral:,.2f}")

# Algumas estatísticas descritivas do faturamento por loja
print("\n--- Estatísticas Descritivas do Faturamento por Loja ---")
print(df_faturamento_total['Faturamento (Preço)'].describe())
```

```python
import matplotlib.pyplot as plt # Para gráficos
import seaborn as sns
```

```python
if not df_faturamento_total.empty:
    plt.figure(figsize=(10, 6))
    sns.barplot(x='Loja', y='Faturamento (Preço)', data=df_faturamento_total, palette='viridis')
    plt.title('Faturamento por Loja (Baseado no Preço do Produto)')
    plt.ylabel('Faturamento (R$)')
    plt.xlabel('Loja')
    plt.xticks(rotation=45) # Rotaciona os nomes das lojas se forem muitos
    # Adicionar os valores nas barras
    for index, row in df_faturamento_total.iterrows():
        plt.text(index, row['Faturamento (Preço)'] + (df_faturamento_total['Faturamento (Preço)'].max() * 0.01), # Posição do texto
                 f"R$ {row['Faturamento (Preço)']:,.0f}", color='black', ha="center")
    plt.tight_layout() # Ajusta o layout para não cortar legendas
    plt.show()
else:
    print("\nNão foi possível gerar o gráfico pois não há dados de faturamento.")

print("\n--- Análise do Faturamento Concluída (baseado na coluna 'Preço') ---")
```

# 2. Vendas por Categoria


```python
def calcular_faturamento_por_categoria(df, nome_da_loja):
  """Calcula o faturamento por categoria de produto para uma loja."""
  if 'Categoria do Produto' in df.columns and 'Preço' in df.columns:
    faturamento_por_categoria = df.groupby('Categoria do Produto')['Preço'].sum()
    print(f"\nFaturamento por categoria da {nome_da_loja}:\n{faturamento_por_categoria}")
    return faturamento_por_categoria
  else:
      print(f"AVISO: Colunas 'Categoria do Produto' ou 'Preço' não encontradas na {nome_da_loja}.")
      return None

faturamento_loja1_cat = calcular_faturamento_por_categoria(loja, "Loja 1")
faturamento_loja2_cat = calcular_faturamento_por_categoria(loja2, "Loja 2")
faturamento_loja3_cat = calcular_faturamento_por_categoria(loja3, "Loja 3")
faturamento_loja4_cat = calcular_faturamento_por_categoria(loja4, "Loja 4")
```

```python
def criar_grafico_barras(dados, titulo):
  """Cria um gráfico de barras a partir dos dados."""
  plt.figure(figsize=(10, 6))
  sns.barplot(x=dados.index, y=dados.values, palette="viridis")
  plt.title(titulo)
  plt.xlabel("Categoria do Produto")
  plt.ylabel("Faturamento Total (R$)")
  plt.xticks(rotation=45, ha='right')
  plt.tight_layout()
  plt.show()
```

```python
criar_grafico_barras(faturamento_loja1_cat, "Faturamento por Categoria - Loja 1")
criar_grafico_barras(faturamento_loja2_cat, "Faturamento por Categoria - Loja 2")
criar_grafico_barras(faturamento_loja3_cat, "Faturamento por Categoria - Loja 3")
criar_grafico_barras(faturamento_loja4_cat, "Faturamento por Categoria - Loja 4")
```

# 3. Média de Avaliação das Lojas

```python
def calcular_faturamento_por_categoria_e_media_avaliacao(df, nome_da_loja):
  """Calcula o faturamento por categoria e a média de avaliação para uma loja."""
  if 'Categoria do Produto' in df.columns and 'Preço' in df.columns and 'Avaliação da compra' in df.columns:
    faturamento_por_categoria = df.groupby('Categoria do Produto')['Preço'].sum()
    media_avaliacao = df['Avaliação da compra'].mean()
    print(f"\nFaturamento por categoria da {nome_da_loja}:\n{faturamento_por_categoria}")
    print(f"Média de avaliação da {nome_da_loja}: {media_avaliacao:.2f}")
    return faturamento_por_categoria, media_avaliacao
  else:
    print(f"AVISO: Colunas 'Categoria do Produto', 'Preço', ou 'Avaliação da compra' não encontradas na {nome_da_loja}. Não foi possível calcular.")
    return None, None
```

```python
def criar_grafico_faturamento_com_media_avaliacao(dados_faturamento, media_avaliacao, titulo):
  """Cria um gráfico de barras de faturamento por categoria com a média de avaliação."""
  plt.figure(figsize=(12, 7))
  sns.barplot(x=dados_faturamento.index, y=dados_faturamento.values, palette="viridis")
  plt.axhline(media_avaliacao, color='red', linestyle='dashed', linewidth=1, label=f'Média Avaliação: {media_avaliacao:.2f}')
  plt.title(titulo)
  plt.xlabel("Categoria do Produto")
  plt.ylabel("Faturamento Total (R$)")
  plt.xticks(rotation=45, ha='right')
  plt.legend()
  plt.tight_layout()
  plt.show()
```

```python
faturamento_loja1_cat, media_avaliacao_loja1 = calcular_faturamento_por_categoria_e_media_avaliacao(loja, "Loja 1")
faturamento_loja2_cat, media_avaliacao_loja2 = calcular_faturamento_por_categoria_e_media_avaliacao(loja2, "Loja 2")
faturamento_loja3_cat, media_avaliacao_loja3 = calcular_faturamento_por_categoria_e_media_avaliacao(loja3, "Loja 3")
faturamento_loja4_cat, media_avaliacao_loja4 = calcular_faturamento_por_categoria_e_media_avaliacao(loja4, "Loja 4")
```

```python
criar_grafico_faturamento_com_media_avaliacao(faturamento_loja1_cat, media_avaliacao_loja1, "Faturamento por Categoria e Média de Avaliação - Loja 1")
criar_grafico_faturamento_com_media_avaliacao(faturamento_loja2_cat, media_avaliacao_loja2, "Faturamento por Categoria e Média de Avaliação - Loja 2")
criar_grafico_faturamento_com_media_avaliacao(faturamento_loja3_cat, media_avaliacao_loja3, "Faturamento por Categoria e Média de Avaliação - Loja 3")
criar_grafico_faturamento_com_media_avaliacao(faturamento_loja4_cat, media_avaliacao_loja4, "Faturamento por Categoria e Média de Avaliação - Loja 4")
```

# 4. Produtos Mais e Menos Vendidos

```python
def identificar_produtos_maior_menor_frete(df, nome_da_loja, top_n=5):
  """Identifica os produtos com maior e menor soma de frete em uma loja."""
  frete_por_produto = df.groupby('Produto')['Frete'].sum()
  maior_frete = frete_por_produto.nlargest(top_n)
  menor_frete = frete_por_produto.nsmallest(top_n)
  print(f"\nProdutos com maior frete total na {nome_da_loja}:\n{maior_frete}")
  print(f"\nProdutos com menor frete total na {nome_da_loja}:\n{menor_frete}")
  return maior_frete, menor_frete
```

```python
def criar_grafico_barras_horizontal(dados, titulo, cor):
  """Cria um gráfico de barras horizontal a partir dos dados."""
  plt.figure(figsize=(10, 6))
  sns.barplot(x=dados.values, y=dados.index, palette=[cor] * len(dados))
  plt.title(titulo)
  plt.xlabel("Soma Total do Frete (R$)")
  plt.ylabel("Produto")
  plt.tight_layout()
  plt.show()
```

```python
for nome_loja, df_loja in lojas_dfs.items():
  maior_frete, menor_frete = identificar_produtos_maior_menor_frete(df_loja, nome_loja)

  # Gráfico dos produtos com maior frete
  criar_grafico_barras_horizontal(maior_frete, f"Top 5 Produtos com Maior Frete Total - {nome_loja}", "lightcoral")

  # Gráfico dos produtos com menor frete
  criar_grafico_barras_horizontal(menor_frete, f"Top 5 Produtos com Menor Frete Total - {nome_loja}", "lightgreen")
```

# 5. Frete Médio por Loja


```python
def calcular_frete_medio(df, nome_da_loja):
  """Calcula o frete médio para uma loja."""
  frete_medio = df['Frete'].mean()
  print(f"O frete médio da {nome_da_loja} é: R$ {frete_medio:.2f}")
  return frete_medio
```

```python
frete_medio_loja1 = calcular_frete_medio(loja, "Loja 1")
frete_medio_loja2 = calcular_frete_medio(loja2, "Loja 2")
frete_medio_loja3 = calcular_frete_medio(loja3, "Loja 3")
frete_medio_loja4 = calcular_frete_medio(loja4, "Loja 4")
```

```python
fretes_medios_por_loja = {
    "Loja 1": frete_medio_loja1,
    "Loja 2": frete_medio_loja2,
    "Loja 3": frete_medio_loja3,
    "Loja 4": frete_medio_loja4
}
```

```python
plt.figure(figsize=(8, 6))
sns.barplot(x=list(fretes_medios_por_loja.keys()), y=list(fretes_medios_por_loja.values()), palette="plasma")
plt.title("Frete Médio por Loja")
plt.xlabel("Loja")
plt.ylabel("Frete Médio (R$)")
plt.tight_layout()
plt.show()
```

# Relatorio

```python
def calcular_faturamento_bruto(df):
  """Calcula o faturamento bruto total (soma dos preços) de uma loja."""
  faturamento_bruto = df['Preço'].sum()
  return faturamento_bruto

def calcular_custo_total_frete(df):
  """Calcula o custo total do frete de uma loja."""
  custo_total_frete = df['Frete'].sum()
  return custo_total_frete
```

```python
faturamento_bruto_loja1 = calcular_faturamento_bruto(loja)
custo_total_frete_loja1 = calcular_custo_total_frete(loja)
lucro_estimado_loja1 = faturamento_bruto_loja1 - custo_total_frete_loja1

faturamento_bruto_loja2 = calcular_faturamento_bruto(loja2)
custo_total_frete_loja2 = calcular_custo_total_frete(loja2)
lucro_estimado_loja2 = faturamento_bruto_loja2 - custo_total_frete_loja2

faturamento_bruto_loja3 = calcular_faturamento_bruto(loja3)
custo_total_frete_loja3 = calcular_custo_total_frete(loja3)
lucro_estimado_loja3 = faturamento_bruto_loja3 - custo_total_frete_loja3

faturamento_bruto_loja4 = calcular_faturamento_bruto(loja4)
custo_total_frete_loja4 = calcular_custo_total_frete(loja4)
lucro_estimado_loja4 = faturamento_bruto_loja4 - custo_total_frete_loja4
```

```python
lucros_estimados = {
    "Loja 1": lucro_estimado_loja1,
    "Loja 2": lucro_estimado_loja2,
    "Loja 3": lucro_estimado_loja3,
    "Loja 4": lucro_estimado_loja4
}
```

```python
loja_menor_lucro = min(lucros_estimados, key=lucros_estimados.get)
menor_lucro = lucros_estimados[loja_menor_lucro]

print("\nFaturamento Bruto e Custo Total de Frete por Loja:")
for loja, lucro in lucros_estimados.items():
  faturamento = calcular_faturamento_bruto(lojas[loja])
  custo_frete = calcular_custo_total_frete(lojas[loja])
  print(f"{loja}: Faturamento Bruto = R$ {faturamento:.2f}, Custo Total Frete = R$ {custo_frete:.2f}, Lucro Estimado = R$ {lucro:.2f}")

print(f"\nA loja com o menor lucro estimado é: {loja_menor_lucro} com um valor de R$ {menor_lucro:.2f}")
```
